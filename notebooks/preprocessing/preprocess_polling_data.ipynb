```python
election_dates = ["2025-01-01", "2019-12-12", "2017-06-08", "2015-05-07", "2010-05-06", "2005-05-05", "2001-06-07",
                    "1997-05-01", "1992-04-09", "1987-06-11", "1983-06-09"]
election_dates = list(reversed(pd.to_datetime(election_dates)))
```

```python
import pandas as pd
import numpy as np

def clean_polling_data(polling, election_dates):
    polling_cleaned = (
        polling.assign(Year=polling['Year'].fillna(method='ffill'))
        .loc[:, ["Year", "Month", "Conservative", "Labour", "LD"]]
        .dropna(subset=["Conservative"])
        .query('Month != "GE"')
        .assign(Month=lambda x: pd.to_datetime(
            np.where(x['Month'].shift() == 'GE', "01 " + x['Month'], "28 " + x['Month']), format="%d %b-%y"))
        .assign(next_election=lambda x: pd.to_datetime(
            election_dates[election_dates.index(next(date for date in election_dates if date > x['Month']))])
        )
        .melt(id_vars=["Year", "Month"], value_vars=["Conservative", "Labour", "LD"], var_name="Party", value_name="Score")
        .groupby(["Party", "next_election"])
    )
    return polling_cleaned

# Example usage:
```

```python
cleaned_data = clean_polling_data(input_polling_df, election_dates)
```
