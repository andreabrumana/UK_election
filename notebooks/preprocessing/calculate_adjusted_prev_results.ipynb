```python
import pandas as pd
```

```python
polling_df = pd.read_csv('/Users/andreabrumana/code/willgreen93/UK_election/raw_data/preprocessed/preprocessed_polling.csv')
polling_df
```

```python
election_results_df = pd.read_csv('/Users/andreabrumana/code/willgreen93/UK_election/raw_data/preprocessed/preprocessed_election_result.csv')
election_results_df
```

```python
elections_polling_df = election_results_df.merge(polling_df, on='year', how='left')
elections_polling_df.info()
```

```python
# Feature engineer polling data at constituency level
elections_polling_df['national_con_pre_ge'].groupby(elections_polling_df['year']).mean()
```

```python
# Feature engineer polling data at constituency level
elections_polling_df['mean_con_share_pre_ge'] = elections_polling_df.groupby('year')['con_share_prev'].transform('mean')
elections_polling_df['mean_lab_share_pre_ge'] = elections_polling_df.groupby('year')['lab_share_prev'].transform('mean')
elections_polling_df['mean_lib_share_pre_ge'] = elections_polling_df.groupby('year')['lib_share_prev'].transform('mean')
elections_polling_df['mean_oth_share_pre_ge'] = elections_polling_df.groupby('year')['oth_share_prev'].transform('mean')
```

```python
elections_polling_df.info()
```

```python
elections_polling_df['con_pre_ge_adjusted'] = elections_polling_df['con_share_prev']/elections_polling_df['mean_con_share_pre_ge']*elections_polling_df['national_con_pre_ge']
elections_polling_df['lab_pre_ge_adjusted'] = elections_polling_df['lab_share_prev']/elections_polling_df['mean_con_share_pre_ge']*elections_polling_df['national_con_pre_ge']
elections_polling_df['lib_pre_ge_adjusted'] = elections_polling_df['lib_share_prev']/elections_polling_df['mean_con_share_pre_ge']*elections_polling_df['national_con_pre_ge']
elections_polling_df['oth_pre_ge_adjusted'] = elections_polling_df['oth_share_prev']/elections_polling_df['mean_con_share_pre_ge']*elections_polling_df['national_con_pre_ge']
```
